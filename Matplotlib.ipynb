```python

import matplotlib.pyplot as plt
import numpy as np

# === LINE PLOT ===
print("Creating line plot...")
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
rainfall = [20, 25, 40, 80, 150, 300]

plt.figure(figsize=(10, 5))
plt.plot(months, rainfall, marker='o', color='blue', linewidth=2)
plt.title('Monthly Rainfall - 2024', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.grid(True, alpha=0.3)
plt.show()

# === SCATTER PLOT ===
print("Creating scatter plot...")
fertilizer = np.array([20, 30, 40, 50, 60, 70, 80])
yield_kg = np.array([2000, 2200, 2600, 2900, 3100, 3200, 3150])

plt.figure(figsize=(10, 5))
plt.scatter(fertilizer, yield_kg, c='green', s=100, alpha=0.7)
plt.title('Fertilizer vs Yield Relationship')
plt.xlabel('Fertilizer (kg/ha)')
plt.ylabel('Yield (kg/ha)')

# Add trend line
z = np.polyfit(fertilizer, yield_kg, 1)
p = np.poly1d(z)
plt.plot(fertilizer, p(fertilizer), "r--", label='Trend')
plt.legend()
plt.show()

# === BAR CHART ===
print("Creating bar chart...")
crops = ['Wheat', 'Rice', 'Cotton', 'Sugarcane']
yields = [2800, 3500, 1500, 8000]
colors = ['gold', 'lightgreen', 'lightblue', 'coral']

plt.figure(figsize=(10, 5))
plt.bar(crops, yields, color=colors, edgecolor='black')
plt.title('Crop Yields Comparison')
plt.xlabel('Crop')
plt.ylabel('Yield (kg/ha)')

# Add value labels
for i, v in enumerate(yields):
    plt.text(i, v + 100, str(v), ha='center')
plt.show()

# === MULTIPLE PLOTS ===
print("Creating subplot...")
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left plot
axes[0].plot(months, rainfall, 'b-o')
axes[0].set_title('Rainfall')
axes[0].set_ylabel('mm')

# Right plot
axes[1].bar(crops, yields, color=colors)
axes[1].set_title('Yields')
axes[1].set_ylabel('kg/ha')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
temperatures = [20, 22, 25, 30, 35, 33]

crops = ["Wheat", "Rice", "Corn", "Sugarcane"]
revenues = [5000, 7000, 4000, 6500]

irrigation = [100, 150, 200, 250, 300]
yield_data = [2, 3, 3.5, 4, 4.5]

land_use_labels = ["Crops", "Livestock", "Storage", "Other"]
land_use_sizes = [50, 20, 15, 15]

# Create 2x2 subplot
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Top-left: Line plot of monthly temperatures
axs[0, 0].plot(months, temperatures, marker='o', color='green')
axs[0, 0].set_title("Monthly Temperatures")
axs[0, 0].set_xlabel("Month")
axs[0, 0].set_ylabel("Temperature (°C)")

# Top-right: Bar chart of crop revenues
axs[0, 1].bar(crops, revenues, color='orange')
axs[0, 1].set_title("Crop Revenues")
axs[0, 1].set_xlabel("Crops")
axs[0, 1].set_ylabel("Revenue")

# Bottom-left: Scatter plot of irrigation vs yield
axs[1, 0].scatter(irrigation, yield_data, color='blue')
axs[1, 0].set_title("Irrigation vs Yield")
axs[1, 0].set_xlabel("Irrigation")
axs[1, 0].set_ylabel("Yield")

# Bottom-right: Pie chart of land use
axs[1, 1].pie(land_use_sizes, labels=land_use_labels, autopct='%1.1f%%')
axs[1, 1].set_title("Land Use Distribution")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
