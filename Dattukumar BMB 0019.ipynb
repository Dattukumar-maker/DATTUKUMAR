```python
#vector operation
import numpy as np
a=np.array([1,6,5,8,4])
b=np.array([5,6,9,2,1])
print(a+b)
```

```python
#martrices operation
x = np.array([
    [3,5,6],
    [3,6,3],
    [7,4,3],
    [4,7,9]
])
print(x)
```

```python
#dot product
result = np.dot([6,7,4],[4,2,9])
print(result)
```

```python
#handling uncertainity
import random
outcomes=["spam","not spam"]
print(random.choice(outcomes))
```

```python
#mean
import numpy as np
a=np.array([1,6,5,8,4])
print(np.mean(a))
```

```python
#normal distribution(bellcurve)
w=7
ir=0.6
for i in range(10):
    w=w*ir
    print(w)
```
