```python

import pandas as pd
import numpy as np

# Create farm data
farm = pd.DataFrame({
    'region': ['North', 'North', 'South', 'South', 'East', 'East'],
    'crop': ['Wheat', 'Rice', 'Wheat', 'Rice', 'Wheat', 'Rice'],
    'area_ha': [10, 8, 12, 6, 9, 7],
    'yield_kg': [25000, 32000, 28000, 24000, 22000, 28000]
})

print("=== Original Data ===")
print(farm)

# === BASIC GROUPBY ===
print("\n=== Average Yield by Crop ===")
by_crop = farm.groupby('crop')['yield_kg'].mean()
print(by_crop)

# === MULTIPLE AGGREGATIONS ===
print("\n=== Multiple Stats by Region ===")
by_region = farm.groupby('region').agg({
    'area_ha': 'sum',
    'yield_kg': ['mean', 'sum', 'count']
})
print(by_region)

# === GROUPBY MULTIPLE COLUMNS ===
print("\n=== By Region AND Crop ===")
multi_group = farm.groupby(['region', 'crop'])['yield_kg'].sum()
print(multi_group)

# === PIVOT TABLE ===
print("\n=== Pivot Table ===")
pivot = farm.pivot_table(
    values='yield_kg',
    index='region',
    columns='crop',
    aggfunc='sum'
)
print(pivot)
```

```python
import pandas as pd

# Example: load your dataset
# df = pd.read_csv("farm_data.csv")

# Assign the 'farm' DataFrame created earlier to 'df'
df = farm.copy()

# -------------------------------
# 1. Group by crop type: average yield
# -------------------------------
avg_yield_by_crop = df.groupby("crop")["yield_kg"].mean().reset_index()
avg_yield_by_crop.rename(columns={"yield_kg": "avg_yield"}, inplace=True)

print("Average Yield by Crop Type:")
print(avg_yield_by_crop)


# -------------------------------
# 2. Group by region: total area and total yield
# -------------------------------
region_summary = df.groupby("region").agg(
    total_area=("area_ha", "sum"),
    total_yield=("yield_kg", "sum")
).reset_index()

print("\nRegion-wise Total Area and Yield:")
print(region_summary)


# -------------------------------
# 3. Pivot table: regions as rows, crops as columns
# -------------------------------
pivot_table = pd.pivot_table(
    df,
    values="yield_kg",
    index="region",
    columns="crop",
    aggfunc="sum",
    fill_value=0
)

print("\nPivot Table (Region vs Crop Yield):")
print(pivot_table)


# -------------------------------
# 4. Best performing region-crop combination
# -------------------------------
best_combo = df.groupby(["region", "crop"])["yield_kg"].sum().reset_index()

best_row = best_combo.loc[best_combo["yield_kg"].idxmax()]

print("\nBest Performing Region-Crop Combination:")
print(best_row)
```
